```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
raw = pd.read_csv('Assocs2.csv')
```

```python
data = raw.copy()
```

```python
data
```

```python
items = []
for item in data.iloc[:,1:].values:
    items.append([i for i in item if type(i)==str])
```

```python
te = TransactionEncoder()
te_array = te.fit(items).transform(items)
df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True, max_len=1)
frequent_itemsets.itemsets = frequent_itemsets.itemsets.apply(lambda x: list(x)[0])
frequent_itemsets
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.bar(x=frequent_itemsets.itemsets, height=frequent_itemsets.support, edgecolor='k')
ax.tick_params(labelcolor='w')
ax.tick_params(axis='x', size=0, labelsize=15, rotation=45)
ax.set_ylabel('Item frequency (relative)\n', fontsize=20, color='w')
plt.show()
```

```python
all_itemsets = apriori(df, min_support=0.00001, use_colnames=True, max_len=1).sort_values('support', ascending=False)
all_itemsets.itemsets = all_itemsets.itemsets.apply(lambda x: list(x)[0])
all_itemsets
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.bar(x=all_itemsets.itemsets, height=all_itemsets.support, edgecolor='k')
ax.tick_params(labelcolor='w')
ax.tick_params(axis='x', size=0, labelsize=15, rotation=45)
ax.set_ylabel('Item frequency (relative)\n', fontsize=20, color='w')
plt.show()
```

```python
all_itemsets = apriori(df, min_support=0.05, use_colnames=True).sort_values('support', ascending=False)
a_rules = association_rules(all_itemsets, metric='confidence', min_threshold=0.5)
a_rules = a_rules[a_rules.consequents.apply(len)==1].reset_index(drop=True)
a_rules
```

```python
a_rules.sort_values('lift', ascending=False).head(10)
```

```python
a_rules[['fish' in a_rules.antecedents[i].union(a_rules.consequents[i]) for i in range(len(a_rules))]].sort_values('lift', ascending=False)
```

```python
a_rules[['fish' in a_rules.antecedents[i].union(a_rules.consequents[i]) for i in range(len(a_rules))]&(a_rules.confidence > 0.8)].sort_values('lift', ascending=False)
```

```python
a_rules[[set(['beer', 'chicken']) <= a_rules.antecedents[i].union(a_rules.consequents[i]) for i in range(len(a_rules))]].sort_values('lift', ascending=False)
```

```python
a_rules[['beer' in a_rules.antecedents[i].union(a_rules.consequents[i]) or 'chicken' in a_rules.antecedents[i].union(a_rules.consequents[i]) for i in range(len(a_rules))]].sort_values('lift', ascending=False)
```
